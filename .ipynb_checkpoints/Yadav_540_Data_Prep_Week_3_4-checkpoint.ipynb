### Activity 3.01 : Geneating statistics from a CSV file

```python
# Load necessary  libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import Boston housing dataset from path = "https://packt.live/2ZPdvc2"
df_csv = pd.read_csv('Boston_housing.csv', sep=',')
```

```python
#check the first 10 records and find the total records

print(f'Total Records: {df_csv.shape}')

print(df_csv.head(10))
```

```python
#Create smaller data frame dropping Chas, nox, b & Lstat
df_csv.drop(['CHAS', 'NOX', 'B', 'LSTAT'],axis=1, inplace = True)
```

```python
#check last 7 records of the new dataframe
df_csv.tail(7)
```

```python
# Plot histogram of all the variables in the new dataframe
df_csv['CRIM'].hist()
plt.show()
df_csv['ZN'].hist()
plt.show()
df_csv['INDUS'].hist()
plt.show()
df_csv['RM'].hist()
plt.show()
df_csv['AGE'].hist()
plt.show()
df_csv['DIS'].hist()
plt.show()
df_csv['RAD'].hist()
plt.show()
df_csv['TAX'].hist()
plt.show()
df_csv['PTRATIO'].hist()
plt.show()
df_csv['PRICE'].hist()
plt.show()
```

```python
# Method 2 printing all histograms using inbuilt pandas hist 
df_csv.hist()
plt.show()
```

```python
# Try to print all plots at once using a for loop. Add unique titles

for column in df_csv:
    df_csv['RAD'].hist()
    plt.title(f'Histogram of variable {column}')
    plt.ylabel(f'Freq of column {column}')
    plt.xlabel(f'{column}')
    plt.show()
```

```python
# Create a scatter plot of crime vs price
df_csv.plot.scatter('CRIM', 'AGE',c='blue',edgecolor='k',s=50)
plt.grid(True)
plt.title('Crime vs Age scatter plot', fontsize = 18)
plt.xlabel('Crime')
plt.ylabel('Age in years')
plt.show()
```

```python
# plot log10(crime) vs Price
df_csv['log_crime'] = np.log10(df_csv['CRIM'])
df_csv.plot.scatter('log_crime', 'AGE',c='blue',edgecolor='k',s=50)
plt.grid(True)
plt.title('Log of Crime vs Age scatter plot', fontsize = 18)
plt.xlabel('Log of Crime')
plt.ylabel('Age in years')
plt.show()
```

```python
# Calculate useful stats
print(f'Mean room {df_csv['RM'].mean()}')
print(f'Median age {df_csv['AGE'].median()}')
print(f'Mean distane {df_csv['DIS'].mean()}')
df_csv['price_20_flag'] = (df_csv['PRICE'] < 20).astype(int)
print(f'Mean price < 20 {df_csv['price_20_flag'].mean()* 100:.2f}%')
```

```python
# Additional means
df_csv.describe()
```

### Activity 4.01: Working with the Adult income Dataset [UCI]

```python
# Load necessary  libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read adult income dataset
# Path = "https://github.com/PacktWorkshops/The-Data-Wrangling-Workshop/blob/master/Chapter04/datasets/adult_income_data.csv"
df = pd.read_csv('adult_income_data.csv', sep= ',')
df.head()
```

```python
# Create a script that will read a text file line by line

with open('adult_income_names.txt',"r") as fd:
    

      # Define initial dataframe
      df_names= []
      for i in fd:
          
           # print(i)
           # Read the lines in a dataframe
           fd.readline()
           df_byline = i.split(':')[0]
           df_names.append(df_byline)
           # print(df_names)
          
df_names
```

```python
# Add a name of *Income* for the response variable to the dataset
df_names.append('Income')
df_names
```

```python
# Assign headings and recreate the file.

df = pd.read_csv('adult_income_data.csv', sep= ',', names=df_names)
print(df.head())


# Find missing values

df.isnull().sum()
```

```python
#create a dataframe with only age , education, and occupation by using subsetting
df_subset = df[['age', 'education', 'occupation']]
df_subset.head(5)
```

```python
#Plot a histogram of age with a bin size of 20
df_subset.hist('age', bins = 20)
```

```python
# Create a function to strip the whitespace characters

def strip_space(col):
    return col.strip()
```

```python
#Use the Apply method to apply this function to all columns with string values, create a new column, copy the values from 
# this new column to the old column, and drop the new column.
df_subset["education_new"] = df_subset["education"].apply(strip_space)
df_subset["education"] = df_subset["education_new"]
df_subset.drop("education_new",axis=1,inplace=True)

df_subset["occupation_new"] = df_subset["occupation"].apply(strip_space)
df_subset["occupation"] = df_subset["occupation_new"]
df_subset.drop("occupation_new",axis=1,inplace=True)

df_subset.head()
```

```python
#Find the number of people aged between 30 and 50
df_subset_age = df_subset[ (df_subset['age'] >= 30) & (df_subset['age']  <=50)]
df_subset_age['age'].count()
```

```python
#group the records based on age and education to find how the mean age is distributed.
df_subset_age.groupby('occupation').mean('age')
```

```python
# group by occupation, show the summary stats of age. find which profession has oldest workers on average and which
# profession has its largest share of the workforce above the 75th percentile.

df_subset_age.groupby('occupation').describe()['age']

# Oldest Workers : Priv-house-serv 40.619048
# LArgest : Exec-managerial 
```

```python
series_1 = pd.Series([7.3, -2.5, 3.4, 1.5],index = ['a', 'c', 'd', 'e'])
series_2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1],index = ['a', 'c', 'e', 'f','g'])
print(series_1)
print(series_2)
```

```python
# Adding Series 1 and Series 2 together gives NaN for indices present in either datasets/ series
print(series_1 + series_2)
```

```python
# Substracting Series 1 and Series 2 together gives NaN for indices present in either datasets/ series
print(series_2 - series_1)
```
