```python
import pandas as pd
```

```python
#Bring in CSV file which is not properly formatted but have comma separated values spread across multiple lines and first line is header
# Use on_bad_lines='skip' to skip lines with too many fields

df = pd.read_csv('MetObjects.csv', on_bad_lines='skip',low_memory=False)
print(f"Shape of dataset is  Rows X Columns : {df.shape}")
df.head(3)
```

```python
# Chapter 7
# Filter out missing data columns with  missing values more than 90% of data

df = df.dropna(axis = 'columns', how = "all") 
df.isna().sum()

```

```python
# Fill in missing data

df = df.fillna(method='bfill')
df = df.fillna(method='ffill')
df.isna().sum()
```

```python
# Chapter 8

# Create hierarchical index
df_index = df.set_index(['Object Number', 'Object ID'])
df_index.index
```

```python

# Pivot the data

# As there are 400K rows, i selected just 2000 rows to pivot post removing duplicates from index. 

df_subset = df.drop_duplicates(subset=['State']).head(2000)
df_pivot = df_subset.pivot(index='State', columns='Region', values='City') # pivoting on Culture and Tags
print(df_pivot.isna().sum())
df_pivot.head()
```

```python
# Chapter 10

# Grouping with Functions
dept_first_letter_counts = df.groupby(df['Department'])['Object ID'].count()
dept_first_letter_counts
```

```python
# Cross Tabs
dept_cross_tab = pd.crosstab(df['Department'], df['Culture'], margins=True)
dept_cross_tab
```

```python
# Chapter 11
# Convert between string and date time

# Convert 'AccessionYear' from string to datetime. At some places it year is in format YYYY-MM-DD and at some places it is just YYYY
# We will convert it to datetime and then back to string with just year and to handle the errors arising due to different formats
# we will use errors='coerce' to convert invalid parsing to NaT (Not a Time)
df['AccessionYear_dt'] = pd.to_datetime(df['AccessionYear'], format="%Y", errors='coerce')

# Convert datetime back to string (year only)
df['AccessionYear_str'] = df['AccessionYear_dt'].dt.strftime('%Y-%m-%d')

df.head()
```
