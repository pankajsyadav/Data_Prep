```python
# Step #1: Import necessary libraries
import pandas as pd
import requests
```

```python
# URL of the JSON data
url = "https://c24215cec6c97b637db6-9c0895f07c3474f6636f95b6bf3db172.ssl.cf1.rackcdn.com/framed/~/media/multimedia/interactives/2015/college_value_added_prod/data/json/list.json"


# Fetch JSON data from URL
response = requests.get(url)
response.raise_for_status()
    
# Parse JSON data
data = response.json()

    
# Convert two_year college data to DataFrame
df = pd.DataFrame(data['two_year'])
df.shape
```

```python
#1 Select relevant columns. In this step we are just bringing in specific columns that we need to, not the whole dataset with 
# other columns like latitude and longitudes
columns_to_keep = [
    'id', 'name', 'city', 'state', 'state_full', 
    'repay', 'sal', 'occ', 'salV', 'occV', 'repayV']


df = df[columns_to_keep]
print(df.shape)
print(df.head())
```

```python
# 2: So the names in CSV are Standardized and wanted to make sure we use the same 
# standardizatoin here to join the data if required text case.

df['name'] = df['name'].str.title()
df['city'] = df['city'].str.title()
df.head()
```

```python
# 3: Convert values from decimals to percentages
pct_columns = ['salV', 'occV', 'repayV']
df[pct_columns] = df[pct_columns] * 100
df[pct_columns] = df[pct_columns].applymap("{:.2f}%".format)
df.head()
```

```python
# 4 Lets create groups based on salary or earning after degree is complete. 
df['Sal_range'] = pd.qcut(df['sal'],q=4,labels=['Low', 'Medium', 'High', 'Rich'])
df.head()
```

```python
#5 Rename columns
df = df.rename(columns={
    'id': 'College ID',
    'name': 'Institution Name',
    'city': 'City',
    'state': 'State',
    'state_full': 'State Full Name',
    'repay': 'Repayment Rate',
    'sal': 'Salary Score',
    'occ': 'Occupational Score',
    'salV': 'Salary Value Added',
    'occV': 'Occupational Value Added',
    'repayV': 'Repayment Value Added',
    'Sal_range': 'Salary Range'
})
```
