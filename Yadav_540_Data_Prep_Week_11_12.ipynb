```python
import sqlite3
import pandas as pd
```

```python
with sqlite3.connect("petsdb") as conn:
    cursor = conn.cursor()
    tables = cursor.execute("PRAGMA table_list")

# check if connection is successful by fetching tables 
table_list = []
for table in tables:
    table_list.append(table)

table_list
```

```python
# count of people belonging to different age groups in the persons table

sqlcnt = "SELECT age, count(*) AS age_count FROM persons GROUP BY age"

counts_rows = cursor.execute(sqlcnt)

get_rows = []
for row in counts_rows:
    get_rows.append(row)
    print(f" We have {row[1]} people aged {row[0]}")
```

```python
# which age group has the maximum number of people
sqlage_max = """ SELECT age, count(*) AS age_count FROM persons GROUP BY age 
              ORDER BY age_count DESC LIMIT 1"""
max_age_row = cursor.execute(sqlage_max) 

for rows in max_age_row:
    print(f" The age group with the maximum number of people is {rows[0]} with {rows[1]} people.")
```

```python
# Count of People do not have last names.
sql_no_last_name = "SELECT id  FROM persons WHERE last_name IS NULL"
no_last_name_count = cursor.execute(sql_no_last_name)
fetch_all_rows = no_last_name_count.fetchall()
print(f"There are {len(fetch_all_rows)} people without last names.")
```

```python
# Cunt of owners with more than one pet
sql_count_owners = """SELECT owner_id, COUNT(*) AS pet_count
FROM pets GROUP BY owner_id HAVING pet_count > 1"""
counts_owners = cursor.execute(sql_count_owners)

# fetch all rows into a list
owners_with_multiple_pets = counts_owners.fetchall()

print(f"Total owners with more than one pet is: {len(owners_with_multiple_pets)}")
```

```python
# Pets that recieved treatment

sqltreat = """SELECT pet_name, treatment_done
FROM pets WHERE treatment_done = 1 """

counts_treated = cursor.execute(sqltreat)

fetch_all_rows = counts_treated.fetchall()

print(f"There are {len(fetch_all_rows)} pets that received treatment.")
```

```python
# pets that received treatment and type of pet is known

sqltreatknown = """SELECT pet_name, pet_type, treatment_done
FROM pets WHERE treatment_done = 1 AND pet_type IS NOT NULL"""

counts_treated = cursor.execute(sqltreatknown)

fetch_all_rows = counts_treated.fetchall()

print(f"There are {len(fetch_all_rows)} pets that received treatment and have a known type.")
```

```python
# Pets from East Port city

# Lets merge the persons and pets tables to get a list of pets from eastport city.
sql2join = """ SELECT pets.pet_name
FROM persons JOIN pets ON persons.id = pets.owner_id 
WHERE persons.city = 'east port'"""
    
counts_treated = cursor.execute(sql2join)

fetch_all_rows = counts_treated.fetchall()

print(f"There are {len(fetch_all_rows)} pets from the East Port city.")
```

```python
# Pets from East Port city that received treatment
sqleastport = """SELECT pet_name, treatment_done
FROM persons JOIN pets ON persons.id = pets.owner_id 
WHERE persons.city = 'east port' AND pets.treatment_done = 1"""
counts_eastport = cursor.execute(sqleastport)
fetch_all_rows = counts_eastport.fetchall()
print(f"There are {len(fetch_all_rows)} pets from East Port city that received treatment.")
```
